### effect of each feature on Campaign Conversion 

#### Age Vs Conversion (Demographic Factor)

```python
# Define age bins and labels
bins = [18, 25, 35, 45, 55, 65, np.inf]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']

# Create age ranges
df['Age Range'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Group by age range and conversion status, then count
age_conversions = df.groupby(['Age Range', 'Conversion'])['Age'].count().unstack(fill_value=0)

x = np.arange(len(age_conversions.index))
width = 0.35
fig, ax = plt.subplots(figsize=(10, 5))
rects1 = ax.bar(x - width/2, age_conversions[0], width, label='Not Converted', color='#66c2a5')
rects2 = ax.bar(x + width/2, age_conversions[1], width, label='Converted', color='#fc8d62')

# Labels and Title
ax.set_xlabel('Age Range')
ax.set_ylabel('Number of Customers')
ax.set_title('Age Range vs Conversion')
ax.set_xticks(x)
ax.set_xticklabels(age_conversions.index)
ax.grid(axis='y', linestyle='--', alpha=0.7)

for rect in rects1 + rects2:
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2, height, f'{height}',
        ha='center', va='bottom', fontsize=10, color='#808080'
    )

ax.legend()
plt.xticks(rotation=0)
plt.show()
```

##### Insights from Age Vs Conversion
<ul>
    <li>35–44 years: Highest conversions</li>
    <li>25–34 years: Second highest</li>
    <li>45–54 years: Third highest</li>
    <li>55–64 years: Fourth highest</li>
    <li>18–24 years: Lower conversions but significant</li>
    <li>65+ years: Lowest conversions</li>
</ul>

## <strong>Return on Ad Spend ROAS</strong>
<ul>
    <li>return on investment of Advertisement campaign on different channels, campaign types</li>
</ul>

a widely referenced benchmark for an “acceptable” ROAS is a 4:1 ratio.

```python
# Define ad labels
ad_labels = ['Low', 'Medium', 'High', 'Very High']
# Create AdSpend bins based on quartiles(0-25, 25-50, 50-75, 75-100)
df['AdSpend_bins'] = pd.qcut(df['AdSpend'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Calculate ROAS for each AdSpend bin
roas = df.groupby('AdSpend_bins', observed=False).apply(
    lambda x: (x['Conversion'].sum() / x['AdSpend'].sum()),
    include_groups=False)

print("ROAS by AdSpend bin: \n",roas)

# Bar chart for ROAS
plt.figure(figsize=(15, 5))
ax = sns.barplot(x=roas.index, y=roas.values, hue=roas.index, palette='Set2')

# Title and labels
plt.title('ROAS vs Ad Spend Range', fontsize=14)
plt.xlabel('Ad Spend Range', fontsize=10)
plt.ylabel('ROAS', fontsize=10)

# bar headings
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 9),
                textcoords='offset points')

# plot markers
plt.plot(roas.index, roas.values, color='seagreen', marker='o', linestyle='-.', linewidth=2)

# Remove grid lines for a cleaner plot
plt.grid(False)

# Show plot
plt.show()
```

## find best k value for cross validation while modeling the data

<p>
<strong>data size analysis for k range</strong>
<ul>
<li>8000 samples is considered a medium-to-large dataset. And the larger dataset generally requires a slightly higher k value to smooth out noise and prevent overfitting.</li>
<li>With 21 features, you may encounter the curse of dimensionality. Higher dimensions tend to require slightly higher k values to counteract noise. Hence, consider testing k = 1 to 60 for better coverage.
</li>
</ul>
</p>

```python
# Finding the best k value
k_values = range(1, 61)  # Testing k values from 1 to 60
cv_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')  # 10-fold cross-validation
    cv_scores.append(scores.mean())

# Determine the optimal k
optimal_k = k_values[np.argmax(cv_scores)]
print(f"Best k value: {optimal_k}")

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(k_values, cv_scores, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('Finding the Optimal k for k-NN')
plt.grid(True)
plt.show()

# Train the model with the best k
best_knn = KNeighborsClassifier(n_neighbors=optimal_k)
best_knn.fit(X_train, y_train)

# Evaluate the model
accuracy = best_knn.score(X_test, y_test)
print(f"Test set accuracy with k={optimal_k}: {accuracy:.2f}")
```

## feature Importance

<p><strong>Why is Feature Importance Important?</strong><br>Feature importance in Machine Learning (ML) refers to techniques used to determine the relevance or contribution of each feature (input variable) in making predictions. It helps identify which features have the most influence on the target variable, guiding feature selection, model optimization, and interpretability.
<ul>
    <li>Improves Model Performance: <br> Focuses on the most informative features, reducing noise and improving accuracy.</li>
    <li>Reduces Overfitting:<br>Eliminates irrelevant or redundant features, lowering complexity and enhancing generalization.</li>
    <li>Enhances Interpretability:<br>Helps explain model predictions, which is crucial.</li>
    <li>Speeds up Computation:<br>Reduces the number of features, leading to faster training and inference times.
</li>
    <li>Guides Feature Engineering:<br>Highlights important features for creating new combinations or transformations.</li>
</ul>
</p>
<p><strong>Methods for Measuring Feature Importance</strong>
<ul>
    <li>Decision Trees and Ensembles: <br>Use Gini Impurity or Information Gain to rank features based on how effectively they split data.</li>
    <li>Random Forest and Gradient Boosting also provide importance scores.</li>
    <li>Feature Importance with PCA (Principal Component Analysis)<br>PCA reduces dimensionality by combining features into principal components. Feature importance can be inferred based on variance explained by each component.</li>
</ul>
</p>

```python
# Calculate feature importances for models that support it

best_models = [(name, model) for name, model in models
               if name not in ("RandomForestClassifier", "AdaBoostClassifier", "DecisionTreeClassifier")]

feature_importances_list = []
model_names_list = []

for model_name, model in best_models:
    feature_importances = None
    if hasattr(model, 'feature_importances_'):
        feature_importances = model.feature_importances_
    elif hasattr(model, 'get_feature_importance'):
        feature_importances = model.get_feature_importance()

    if feature_importances is not None:
        feature_importances_list.append(feature_importances)
        model_names_list.append(model_name)
print("feature_importances_list: ", feature_importances_list)
print("model_names_list: ",model_names_list)
```

```python
# Ensure X_train is a DataFrame

X_train1 = pd.DataFrame(X_train, columns=['Gender_Male', 'CampaignChannel_PPC', 'CampaignChannel_Referral',
       'CampaignChannel_SEO', 'CampaignChannel_Social Media',
       'CampaignType_Consideration', 'CampaignType_Conversion',
       'CampaignType_Retention', 'Age', 'Income', 'AdSpend',
       'ClickThroughRate', 'ConversionRate', 'WebsiteVisits', 'PagesPerVisit',
       'TimeOnSite', 'SocialShares', 'EmailOpens', 'EmailClicks',
       'PreviousPurchases', 'LoyaltyPoints'])
if not isinstance(X_train1, pd.DataFrame):
    # Filter models with feature importance support
    best_models = [(name, model) for name, model in models if hasattr(model, 'feature_importances_') or hasattr(model, 'coef_')]

feature_importances_list = []
model_names_list = []

# Extract feature importances
for model_name, model in best_models:
    if hasattr(model, 'feature_importances_'):
        feature_importances = model.feature_importances_
    elif hasattr(model, 'coef_'):
        feature_importances = np.abs(model.coef_).ravel()  # Handle coefficients
    else:
        continue

    feature_importances_list.append(feature_importances)
    model_names_list.append(model_name)

# Plot feature importances
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(20, 30))
axes = axes.flatten()

for i in range(min(8, len(feature_importances_list))):
    # Create DataFrame for importances
    feature_importances_df = pd.DataFrame({'Feature': X_train1.columns, 'Importance': feature_importances_list[i]})
    feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

    # Plot top 10 features
    sns.barplot(x='Importance', y='Feature', data=feature_importances_df[:10], ax=axes[i], palette="viridis")
    axes[i].set_title(f"Top 10 Features - {model_names_list[i]}", fontweight='bold')
    axes[i].set_xlabel('Importance', fontsize=12)
    axes[i].set_ylabel('Feature', fontsize=12)
    axes[i].grid(False)
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)

# Hide unused axes
for i in range(len(feature_importances_list), len(axes)):
    axes[i].set_visible(False)

plt.tight_layout(pad=3.0)
plt.show()
```

## Hyper parameter tuning

<p><strong>Importance of Hyperparameter Tuning Using Cross-Validation</strong><br>Hyperparameter tuning is a crucial step in machine learning model development because it optimizes model performance by finding the best combination of hyperparameters. When combined with cross-validation, it ensures that the results are reliable and generalizable to unseen data.<br>
    <p>Hyperparameter tuning with cross-validation is vital because it provides an objective evaluation of model performance. It avoids the pitfalls of overfitting or underfitting and ensures that the chosen model is generalizable to unseen data. This process leads to building robust and high-performing machine learning models suitable for real-world applications.</p><br>
    
<ul>
    <li><strong>Hyperparameters</strong> are configuration settings that govern the training process and are not learned from the data (e.g., learning rate, number of trees in Random Forest, or number of hidden layers in neural networks).
<strong>Tuning</strong> involves searching for the best values for these parameters to improve model performance.</li>
    <li><strong> Why Combine Cross-Validation with Hyperparameter Tuning?</strong><br><strong>Avoids Overfitting:</strong><br>Cross-validation splits the data into multiple training and validation sets, ensuring the model is evaluated on different subsets of data.<br>Prevents the model from being overly optimized to a single training set, reducing overfitting. </li>
    <li><strong>Balances Bias and Variance:</strong><br>Cross-validation provides insights into whether the model suffers from high bias (underfitting) or high variance (overfitting), guiding hyperparameter adjustments accordingly. </li>
    <li><strong>Optimizing Performance:</strong><br> Ensures that the model achieves its maximum predictive accuracy.</li>
    <li><strong>Better Resource Management: </strong><br> Helps balance computational cost by selecting optimal complexity.</li>
    <li><strong>Reliable Comparison of Models:</strong><br> Reliable Comparison of Models: Allows fair evaluation of different models or configurations.</li>

</ul>
</p>

```python
#Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Iterate through models
for name, model in (dict(models).items()):
    print(f"Model: {name}")
    print("="*50)
    
    # Calculate metrics using cross-validation
    precision = cross_val_score(model, X_train, y_train, cv=cv, scoring=make_scorer(precision_score, average='weighted'))
    recall = cross_val_score(model, X_train, y_train, cv=cv, scoring=make_scorer(recall_score, average='weighted'))
    f1 = cross_val_score(model, X_train, y_train, cv=cv, scoring=make_scorer(f1_score, average='weighted'))
    accuracy = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')

    # Display results
    print(f"Accuracy: {np.mean(accuracy):.2f}")
    print(f"Precision: {np.mean(precision):.2f}")
    print(f"Recall: {np.mean(recall):.2f}")
    print(f"F1-Score: {np.mean(f1):.2f}")
    print("\n")
```

### ROC- AUC Curve


<p>
    <strong>Importance of the ROC-AUC Curve</strong>
<ul>
    <li>The ROC-AUC curve is an essential tool for evaluating classification models</li>
    <li>The ROC-AUC curve (Receiver Operating Characteristic - Area Under Curve) is a key evaluation metric for classification models, especially in scenarios involving imbalanced datasets.
</li>
    <li><strong>Evaluates Model Discrimination:</strong><br>Measures how well the model separates positive and negative classes.
Useful in cases of imbalanced datasets, unlike accuracy which can be misleading.</li>
    <li><strong>Compares Multiple Models:</strong><br>Enables easy comparison of different models based on their AUC scores.
Higher AUC implies a better-performing model</li>
    <li><strong>ROC Curve Shape</strong><br>Shows the model's ability to distinguish between classes across thresholds.</li>
    <li><strong>AUC Score</strong><br>Provides a single value summarizing performance—higher values indicate better performance.</li>
    <li><strong>Diagonal Line (Random Guess)</strong><br>Helps visually compare how far the model performs above chance-level classification.</li>
</ul>
</p>

```python
# Create subplots dynamically based on number of models
n_models = len(models)
n_rows = (n_models + 2) // 3  # Arrange 3 plots per row
fig, axes = plt.subplots(n_rows, 3, figsize=(20, 5 * n_rows))

# Flatten axes if more than 1 row, to avoid indexing errors
axes = axes.flatten() if n_rows > 1 else axes

for ax, (name, model) in zip(axes, (dict(models).items())):
    # Train the model
    model.fit(X_train, y_train)

    # Predict probabilities for the positive class
    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_test)[:, 1]
    else:
        y_score = model.decision_function(X_test)

    # Calculate ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve in each subplot
    ax.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
    ax.plot([0, 1], [0, 1], color='grey', linestyle='--')  # Diagonal line
    ax.set_title(name, fontsize=12)
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.legend(loc='lower right')
    ax.grid(True)

# Hide empty subplots if models < total axes
for i in range(n_models, len(axes)):
    fig.delaxes(axes[i])

# Final adjustments and display plot
plt.tight_layout(pad=3.0)
plt.show()
```

```python
import pandas as pd
import numpy as np
import joblib
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Load the dataset
cust_conv = pd.read_csv("digital_marketing_campaign_dataset.csv")

# Encode categorical columns using pd.get_dummies()
encoding_cols = ['Gender', 'CampaignChannel', 'CampaignType']
encoded_data_df = pd.get_dummies(cust_conv, columns=encoding_cols, drop_first=True)

# Separate features and target variable
X = encoded_data_df.drop(["Conversion", "AdvertisingPlatform", "AdvertisingTool"], axis=1)
y = encoded_data_df["Conversion"]

# Handle imbalance using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Scale the features
scaler = StandardScaler()
X_resampled = scaler.fit_transform(X_resampled)

# Train initial XGBoost model on resampled data
model = XGBClassifier(random_state=42)
model.fit(X_resampled, y_resampled)

# Get feature importances from the trained model
importances = model.feature_importances_

# Identify the top 5 features based on importance
top_features_idx = np.argsort(importances)[::-1][:5]  # Get indices of top 5 features
top_features = X.columns[top_features_idx]

# Select only the top 5 features
X_top5 = X[top_features]

# Apply SMOTE to the reduced features
X_top5_resampled, y_top5_resampled = smote.fit_resample(X_top5, y)

# Train the XGBoost model with the top 5 features
model_top5 = XGBClassifier(random_state=42)
model_top5.fit(X_top5_resampled, y_top5_resampled)

# Save the trained model using joblib
joblib.dump(model_top5, 'xgb_model_top5_features101.joblib')

# Save the scaler for deployment purposes
joblib.dump(scaler, 'scaler.joblib')

# Save the selected top 5 features
joblib.dump(top_features, 'top_features.joblib')

print("Model, scaler, and top features saved successfully!")
```
